# Text Splitter

### CharacterTextSplitter

```python
text = data[1].page_content[:1500]
```

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size = 200,
    chunk_overlap=0
)
```

```python
chunks = splitter.split_text(text)
chunks
```

```python
len(chunks)
```

```python
for chunk in chunks:
    print(len(chunk))
```

### RecursiveTextSplitter

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " "],
    chunk_size=200,
    chunk_overlap=0,
    length_function = len
)
```

```python
chunks = r_splitter.split_text(text)
for chunk in chunks:
    print(len(chunk))
```

# FAISS TUTORIAL

```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv("sentiment_tweets3.csv")
df.shape
```

```python
df
```

### Step 1 : Create source embeddings for the text column

```python
from sentence_transformers import SentenceTransformer
```

```python
df = df[:100]
df
```

```python
encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df["message to examine"])
vectors
```

```python
vectors.shape
```

```python
dim = vectors.shape[1]
dim
```

#### Step 2 : Build a FAISS Index for vectors

```python
import faiss

index = faiss.IndexFlatL2(dim)
```

#### Step 3 : Normalize the source vectors (as we are using L2 distance to measure similarity) and add to the index

```python
index.add(vectors)
```

```python
index
```

#### Step 4 : Encode search text using same encorder and normalize the output vector

```python
search_query = "Not in a good mood today"
vec = encoder.encode(search_query)
vec.shape
```

```python
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
```

```python
faiss.normalize_L2(svec)
```

#### Step 5: Search for similar vector in the FAISS index created

```python
distances, I = index.search(svec, k=2)
distances
```

```python
row_indices = I.tolist()[0]
row_indices
```

```python
df.loc[row_indices]
```

```python
search_query
```

## RETRIEVAL

```python
import os
import streamlit as st
import pickle
import time
import langchain
from langchain import OpenAI
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.qa_with_sources.loading import load_qa_with_sources_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import UnstructuredURLLoader
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from dotenv import load_dotenv
```

```python
#load openAI api key
load_dotenv()
OPENAI_API_KEY = os.environ["OPENAI_API_KEY"]
```

```python
# Initialize LLM with required parameters
# model_name="gpt-3.5-turbo"
model_name="gpt-3.5-turbo-instruct"
llm = OpenAI(model=model_name, temperature=0.5)
```

```python
## LOAD DATA
loaders = UnstructuredURLLoader(
    urls=[
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html",
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html"
]
)
data = loaders.load()
len(data)
```

```python
## SPLIT DATA TO CREATE CHUNKS
text_spitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap =200)
```

```python
# As data is of type documents we can directly use split_documents over split_text in order to get the chunks.
docs = text_spitter.split_documents(data)
docs
```

```python
len(docs)
```

```python
docs[0]
```

```python
## CREATE EMBEDDINGS FOR THESE CHUNKS AND SAVE THEM TO FAISS INDEX  
# Create the embeddings of the chunks using openAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")

# Pass the documents and embeddings inorder to create FAISS vector index
vector_index_openai = FAISS.from_documents(docs, embeddings)
```

```python
# Storing vector index created in local
file_path = "vector_index.pkl"
with open(file_path, "wb") as f:
    pickle.dump(vector_index_openai, f)
```

```python
if os.path.exists(file_path):
    with open(file_path, "rb") as f:
        vectorIndex = pickle.load(f)
```

```python
## RETRIEVE SIMILAR EMBEDDINGS FOR A GIVEN QUESTION AND CALL LLM TO RETRIEVE FINAL ANSWER
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorIndex.as_retriever())
```

```python
query = "what is the price of Tiago iCNG?"

langchain.debug = True

chain({"question":query}, return_only_outputs=True)
```
